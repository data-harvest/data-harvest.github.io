```python
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
gpd_map = gpd.read_file("sjc_croplyr2020.shp")
```

```python
gpd_map.head(2)
```

```python
# set range for the colors
vmin = 0
vmax = 100
vstep = 10
# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(20, 20))

#Use cmap = 'BuGn' to get nice color-shaded plot
gpd_map.plot(column='calc_acres', cmap='Reds',
             linewidth=0.8, ax=ax, edgecolor='black')


sm = plt.cm.ScalarMappable(cmap='Reds', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)

plt.title('size')
plt.axis('off')
plt.show()
```

```python
partial_gpd = gpd_map.cx[6400000:6450000, 2100000:2150000]
# set range for the colors
vmin = 0
vmax = 100
vstep = 10
# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(20, 20))

#Use cmap = 'BuGn' to get nice color-shaded plot
partial_gpd.plot(column='calc_acres', cmap='Reds',
             linewidth=0.8, ax=ax, edgecolor='black')


sm = plt.cm.ScalarMappable(cmap='Reds', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)

plt.title('size')
plt.axis('off')
plt.show()
```
